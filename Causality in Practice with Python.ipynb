```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n = 100
study_hours = np.random.uniform(1, 10, n)
exam_scores = 5 * study_hours + np.random.normal(0, 5, n)

# Convert to DataFrame
data = pd.DataFrame({'StudyHours': study_hours, 'ExamScores': exam_scores})

# Scatter plot to visualize the relationship
plt.scatter(data['StudyHours'], data['ExamScores'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs. Exam Scores')
plt.show()

# Fit a linear regression model
X = sm.add_constant(data['StudyHours'])  # Add constant term for intercept
model = sm.OLS(data['ExamScores'], X).fit()

# Print the summary of the regression model
print(model.summary())

# Plot the regression line
plt.scatter(data['StudyHours'], data['ExamScores'], label='Data Points')
plt.plot(data['StudyHours'], model.predict(X), color='red', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs. Exam Scores with Regression Line')
plt.legend()
plt.show()
```
