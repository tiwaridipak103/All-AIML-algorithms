```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of predictions
y_true = 1  # True class label
preds = np.linspace(-2, 2, 400)

# Calculate the losses
mse_loss = (preds - y_true) ** 2
log_loss = np.log(1 + np.exp(-y_true * preds))
hinge_loss = np.maximum(0, 1 - y_true * preds)

# Calculate the 0-1 loss
zero_one_loss = np.where(y_true * preds <= 0, 1, 0)

# Plot the losses
plt.figure(figsize=(10, 4))

plt.plot(preds, mse_loss, label='Mean Squared Error (MSE)', color='blue')
plt.plot(preds, log_loss, label='Logistic Loss', color='red')
plt.plot(preds, hinge_loss, label='Hinge Loss', color='green')
plt.plot(preds, zero_one_loss, label='0-1 Loss', color='purple')


plt.xlabel('Prediction Error')
plt.ylabel('Loss')
plt.title('Loss Function Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of predicted probabilities
y_pred = np.linspace(0.001, 0.999, 1000)

# Compute logistic loss for y = 1 and y = 0
loss_y1 = -np.log(y_pred)
loss_y0 = -np.log(1 - y_pred)

# Plot the logistic loss functions
plt.figure(figsize=(10, 6))
plt.plot(y_pred, loss_y1, label='Loss when y=1', color='blue')
plt.plot(y_pred, loss_y0, label='Loss when y=0', color='red')
plt.xlabel('Predicted Probability ()')
plt.ylabel('Logistic Loss')
plt.title('Logistic Loss vs. Predicted Probability')
plt.legend()
plt.grid(True)
plt.show()
```
