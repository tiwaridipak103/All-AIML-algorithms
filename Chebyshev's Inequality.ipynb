```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Apply Chebyshev's inequality
k = 2
lower_bound = mean - k * std_dev
upper_bound = mean + k * std_dev

# Calculate the proportion of data within the bounds
within_bounds = np.sum((data >= lower_bound) & (data <= upper_bound)) / len(data)
chebyshev_bound = 1 - 1 / k**2

print(f"Proportion of data within {k} standard deviations: {within_bounds:.4f}")
print(f"Chebyshev's inequality bound: {chebyshev_bound:.4f}")

# Plotting
plt.hist(data, bins=30, alpha=0.6, color='g')
plt.axvline(lower_bound, color='r', linestyle='dashed', linewidth=2)
plt.axvline(upper_bound, color='r', linestyle='dashed', linewidth=2)
plt.title('Data Distribution with Chebyshev Bounds')
plt.show()
```
