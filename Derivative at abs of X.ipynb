```python
import numpy as np

# Define the absolute value function
def f(x):
    return abs(x)

# Function to calculate the numerical derivative using a small step size h
def numerical_derivative(f, x, h=1e-6):
    return (f(x + h) - f(x)) / h

# Define the point where we want to evaluate the derivatives
x0 = 0.0

# Calculate left-hand derivative (from the negative side)
f_left_derivative = numerical_derivative(f, x0 - 1e-6)

# Calculate right-hand derivative (from the positive side)
f_right_derivative = numerical_derivative(f, x0 + 1e-6)

print(f"Left-hand derivative at x = {x0}: {f_left_derivative}")
print(f"Right-hand derivative at x = {x0}: {f_right_derivative}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the absolute value function
def f(x):
    return np.abs(x**2 + 3*x)


# Generate x values
x = np.linspace(-3.5, 0.5, 400)  # Adjusted range to show the behavior around x = -3

# Compute corresponding y values
y = f(x)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = |x^2 + 3x|', color='blue')
plt.title('Plot of y = |x^2 + 3x|')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)  # horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5)  # vertical line at x=0
plt.grid(True)
plt.legend()

# Highlight the point of non-differentiability at x=-3
plt.scatter([-3], [0], color='red', label='Non-differentiable at x=-3', zorder=5)
plt.legend()

plt.show()

```
